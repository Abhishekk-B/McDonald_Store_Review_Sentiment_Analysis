```python
def text_preprocessing(doc):
    tokens=doc.split()
    words=[token.strip() for token in tokens]
    short_form = short_forms()
    converted = [short_form[word] if word in short_form else word for word in words]
    data = " ".join(converted)
    text=re.sub(r'[^a-bA-z\s]',"",data.lower())
    return text
```

```python
df['processed_text']=df.review.apply(lambda x:text_preprocessing(x))
df.head()
```

```python
import spacy
nlp=spacy.load("en_core_web_lg")
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import WordNetLemmatizer
lm=WordNetLemmatizer()
nlp.Defaults.stop_words-={'not','never'}
```

```python
def cleaned_text(text):
    filtered_word=[]
    doc=nlp(text)
    for word in doc:
        if not (word.is_stop) and not (word.is_punct):
            filtered_word.append(lm.lemmatize(str(word)))    
    text=[word for word in filtered_word if len(word)>1]
    text=" ".join(text)
    return text    
```

```python
df['cleaned_text']=df['processed_text'].apply(lambda x:cleaned_text(x))
df_copy=df.copy()
```

### Performing Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
sentiments = []
for review in df['cleaned_text']:
    sentiment = sia.polarity_scores(review)
    sentiments.append(sentiment)
```

```python
sentiments
```

```python
df_temp=pd.DataFrame(sentiments)
```

```python
df_new=df.join(df_temp)
```

```python
df=df_new[['cleaned_text','rating','neg','neu','pos','compound']]
```

```python
df.head()
```

```python
sentiment_labels = []
for index, row in df.iterrows():
    if row['compound']>=0.05:
        if (row['rating']=='1 star'):
            sentiment_labels.append('Negative')
        else:
            sentiment_labels.append('Positive')
    elif row['compound']<=-0.05:
        if (row['rating']=='5 stars'):
            sentiment_labels.append('Positive')
        else:
            sentiment_labels.append('Negative')
    else:
        sentiment_labels.append('Neutral')
```

```python
df['sentiment'] = sentiment_labels
df[['cleaned_text', 'sentiment']]
```

```python
df.sentiment.value_counts()
```

```python
X = df['cleaned_text']
y = df['sentiment']
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
v = TfidfVectorizer()
X_train = v.fit_transform(X_train)
X_test = v.transform(X_test)
```

```python
len(v.vocabulary_)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
model1=LogisticRegression(max_iter=400)
model1.fit(X_train,y_train)
y_pred = model1.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
model2=MultinomialNB()
model2.fit(X_train,y_train)
y_pred = model2.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
model3=KNeighborsClassifier()
model3.fit(X_train,y_train)
y_pred = model3.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
model4=SVC()
model4.fit(X_train,y_train)
y_pred = model4.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
model5=RandomForestClassifier()
model5.fit(X_train,y_train)
y_pred = model5.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
def predict_sentiment(review):
    review_tfidf = v.transform([review])
    sentiment = model4.predict(review_tfidf)
    return sentiment[0]
```

```python
new_review = "This restaurant provide a very good service."
predicted_sentiment = predict_sentiment(new_review)
predicted_sentiment=lb.inverse_transform([predicted_sentiment])
print("Predicted sentiment:", predicted_sentiment[0])
```

```python
new_review = "The food is not so good. The taste are so pathetic."
predicted_sentiment = predict_sentiment(new_review)
predicted_sentiment=lb.inverse_transform([predicted_sentiment])
print("Predicted sentiment:", predicted_sentiment[0])
```
